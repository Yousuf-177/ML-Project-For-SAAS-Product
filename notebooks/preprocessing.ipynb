### **Import Libraries & Functions**

```python
import sys
import pandas as pd
from pathlib import Path
```

```python
PROJECT_ROOT = Path(r"D:\Project").resolve()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

from utils.features import numerical_cols, categorical_cols, target_col, remove_cols
from utils.utils import load_data, split_data_stratified
from utils.utils  import PreprocessorManager
```

### **Read Interim Data**

```python
df = load_data("raw/saas_lead.csv")
df.head(3)
```

### **Remove Columns**

```python
df = df.drop(columns=remove_cols)
df.shape
```

### **Split Data**

```python
TARGET_COL = target_col

X = df.drop(columns=[TARGET_COL])
y = df[TARGET_COL]

X_train, X_test, y_train, y_test = split_data_stratified(
    X, y, test_size=0.2, random_state=42
)
```

### **Preprocess Training Data**

```python
# Read numerical and categorical columns from features.py
NUM_COL = numerical_cols
CAT_COL = categorical_cols
```

```python
# Processes TRAIN
preprocessor = PreprocessorManager(processor_path=PROJECT_ROOT / "utils" / "processor.pkl")
X_train_processed = preprocessor.fit_transform_train(
    X_train,
    numerical_cols=NUM_COL,
    categorical_cols=CAT_COL
)
```

```python
# Processes TEST
X_test_processed = preprocessor.transform_test(X_test)
```

### **Save Training and Testing Data**

```python
X_train_processed.to_csv(PROJECT_ROOT / "data/processed/X_train.csv", index=False)
X_test_processed.to_csv(PROJECT_ROOT / "data/processed/X_test.csv", index=False)
y_train.to_csv(PROJECT_ROOT / "data/processed/y_train.csv", index=False)
y_test.to_csv(PROJECT_ROOT / "data/processed/y_test.csv", index=False)
```
